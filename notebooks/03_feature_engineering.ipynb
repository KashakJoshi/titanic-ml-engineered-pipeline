# Phase 3 – Advanced Feature Engineering

In this phase, we:

- Extract Title
- Create Family Features
- Extract Deck
- Add AgeGroup
- Add FareGroup
- Add Class × Sex interaction
- Handle Missing Values
- Drop Unused Columns
- Encode Categorical Features

All transformations are handled inside src/feature_engineering.py

```python
import sys
sys.path.insert(0, "../src")

import pandas as pd

from feature_engineering import (
    extract_title,
    create_family_features,
    extract_deck,
    add_age_group,
    add_fare_group,
    add_class_sex_feature,
    handle_missing_values,
    drop_unused_columns,
    encode_features
)
```

```python
train_df = pd.read_csv("../data/raw/train.csv")
train_df.head()
```

###  Final Feature Engineering Pipeline (Test Data)


```python
import sys
import os

sys.path.append(os.path.abspath(".."))
```

```python
from src.feature_engineering import (
    extract_title,
    create_family_features,
    extract_deck,
    handle_missing_values,
    drop_unused_columns,
    encode_features,
    add_age_group,
    add_fare_group,
    add_class_sex_feature
)
```

```python

test = pd.read_csv("../data/raw/test.csv")
print(test.shape)
```

```python
def preprocess_test(df):
    df = extract_title(df)
    df = create_family_features(df)
    df = extract_deck(df)
    df = handle_missing_values(df)
    df = add_age_group(df)
    df = add_fare_group(df)
    df = add_class_sex_feature(df)
    df = drop_unused_columns(df)
    df = encode_features(df)
    return df

test_processed = preprocess_test(test)
```
